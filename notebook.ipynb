```python
# Pip YOLOv8
!pip install ultralytics
import sympy
from ultralytics import YOLO

data_yaml = """
train: /content/playing_cards/train/images
val: /content/playing_cards/valid/images

nc: 52
names: ['10c', '10d', '10h', '10s', '2c', '2d', '2h', '2s', '3c', '3d', '3h', '3s', '4c', '4d', '4h', '4s', '5c', '5d', '5h', '5s', '6c', '6d', '6h', '6s', '7c', '7d', '7h', '7s', '8c', '8d', '8h', '8s', '9c', '9d', '9h', '9s', 'Ac', 'Ad', 'Ah', 'As', 'Jc', 'Jd', 'Jh', 'Js', 'Kc', 'Kd', 'Kh', 'Ks', 'Qc', 'Qd', 'Qh', 'Qs']
"""

with open('/content/playing_cards/data.yaml', 'w') as f:
    f.write(data_yaml)

model = YOLO('yolov8n.pt')
model.train(data='/content/playing_cards/data.yaml', epochs=10, imgsz=640)

# Save model
model.export(format='torchscript')
```

```python
# Evaluate on test set
results = model.val(data='/content/playing_cards/data.yaml')

# Print mAP50 and mAP50-95
print(f"mAP50: {results.box.map50:.2f}")  # Mean Average Precision @ IoU=50
print(f"mAP50-95: {results.box.map:.2f}")  # Mean Average Precision @ IoU=50:95
```
